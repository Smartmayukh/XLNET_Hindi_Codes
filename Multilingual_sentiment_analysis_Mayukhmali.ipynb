```
import nltk
nltk.download('stopwords')
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
sw =set(stopwords.words('english'))
ps=PorterStemmer()
```

```
def clean_text(sample):
    sample=sample.lower()
    sample=re.sub("[^a-zA-Z]+"," ",sample)
    sample = sample.split()
    sample=[ ps.stem(s) for s in sample if s not in sw]
    sample= " ".join(sample)
    return sample
```

```
import csv 
import pandas as pd
from google.colab import files
uploaded = files.upload()
update_data_hindi = pd.read_csv('TransEnglish.csv')
update_data_hindi['text']=update_data_hindi['text'].apply(clean_text)
update_data_hindi=update_data_hindi[update_data_hindi['polarity']!='conflict']
update_data_hindi.drop('Unnamed: 0', inplace=True, axis=1)
update_data_hindi.reset_index(drop=True, inplace=True)
update_data_hindi.head(2000)
```

```
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
y=update_data_hindi['polarity'].values
y=label_encoder.fit_transform(y)
y[:100]
```

```
from sklearn.feature_extraction.text import CountVectorizer ,TfidfTransformer
import tensorflow as tf
corpus=update_data_hindi['text'].values
cv = CountVectorizer(max_df=0.5)
X=cv.fit_transform(corpus)
tfidf=TfidfTransformer()
X=tfidf.fit_transform(X)
print(X[0])
```

```
from tensorflow import keras
import tensorflow
from keras import models
from keras.layers import Dense
from keras.layers import LeakyReLU
from keras.layers import LSTM
from keras.layers import Dropout
from tensorflow.keras.layers import BatchNormalization
model=models.Sequential()
model.add( Dense(128,input_shape=(X.shape[1],)))
model.add(LeakyReLU(alpha=0.05))
model.add(BatchNormalization(axis=-1,
    momentum=0.99,
    epsilon=0.001,
    center=True,
    scale=True,
    beta_initializer="zeros",
    gamma_initializer="ones",
    moving_mean_initializer="zeros",
    moving_variance_initializer="ones",
    beta_regularizer=None,
    gamma_regularizer=None,
    beta_constraint=None,
    gamma_constraint=None,
    renorm=False,
    renorm_clipping=None,
    renorm_momentum=0.99,
    fused=None,
    trainable=True,
    virtual_batch_size=None,
    adjustment=None,
    name=None,
    ))
model.add( Dense(64))
model.add(LeakyReLU(alpha=0.05))
model.add( Dense(32))
model.add(Dropout(0.5))
model.add(LeakyReLU(alpha=0.05))
model.add( Dense(16))
model.add(Dropout(0.5))
model.add(LeakyReLU(alpha=0.05))
model.add( Dense(1,activation="sigmoid"))
model.compile(optimizer='adam', loss="binary_crossentropy", metrics=['accuracy'])
model.summary()
```

```
import tensorflow
model.compile(optimizer='adam', loss="binary_crossentropy", metrics=['accuracy'])
X_val=X[:200]
X_train=X[200:]
y_val=y[:200]
y_train=y[200:]
X_val.sort_indices()
X_train.sort_indices()
y_val
y_train
```

```
hist=model.fit(X_train,y_train,batch_size=20,epochs=7,validation_data=(X_val,y_val))
```

```
import matplotlib.pyplot as plt
result=hist.history
plt.plot(result['val_accuracy'], label="Val acc")
plt.plot(result['accuracy'],label="Train acc")
plt.legend()
plt.show()
```
